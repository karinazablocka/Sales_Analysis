# Sales Analysis

#### Import Necessary Libraries

```python
import pandas as pd
import os
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

### TASK #1: Merging 12 months of sales data into a single file

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

data={}

for m in months:
    d = pd.read_csv("./Sales_Data/Sales_{}_2019.csv".format(m))
    data[m] = d
    
df = pd.concat(data, ignore_index=True)
df = df.sort_values('Order Date')
```

#### or

```python
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame() # Empty DF

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("all_data.csv", index=False) # Save concatenated DFrames in one file
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

## Augment data with additional columns

### TASK #2: Add Month Column

#### Delete null values

```python
all_data.info()
```

```python
null = all_data[all_data["Order ID"].isnull()]
null.head()
```

```python
all_data = all_data.dropna(how='all')
```

```python
all_data.isnull().sum()
```

#### Delete rows with "Order ID" in Order ID column

```python
all_data["Order ID"].value_counts()
```

```python
all_data = all_data[all_data["Order ID"]!="Order ID"]
```

#### Add new column


```python
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype("int")

all_data.head()
```

### Queastion 1: What was the best moth for sales? How much was earned that month?

#### Add new column Quantity Ordered * Price Each = ["Profit"]

```python
all_data["Quantity Ordered"] = all_data["Quantity Ordered"].astype("int")
all_data["Price Each"] = all_data["Price Each"].astype("float")

all_data.info()
```

```python
all_data["Profit"] = all_data["Quantity Ordered"] * all_data["Price Each"]
```

```python
all_data.head(10)
```

```python
grouped = all_data.groupby(all_data["Month"]).sum()
```

```python
grouped
```

```python
fig, ax = plt.subplots(1,1, figsize=(8, 6), sharex=False)

#my_colors = [(x/12, x/17.0, 0.75) for x in range(len(grouped))] # <-- Quick gradient example along the Red/Green dimensions.
my_colors = ['#644655', '#79525f', '#855E6F', '#9E768F','#9F829D','#9F8DAA','#9FA4C4','#A4AFC8','#A9B9CB','#B3CDD1','#BDDFC7','#C7F0BD']

ax.bar(grouped.index, grouped["Profit"], color=my_colors) 
#grouped["Profit"].plot(kind="bar", colormap='Dark2')

ax.set_xticks(np.arange(len(grouped)+1))
ax.ticklabel_format(axis="y", style="plain", scilimits=(0,0))
ax.set_title("Monthly Profit", fontsize=14)
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Profit", fontsize=12)


plt.show()
```

### Question 4: What products are most often sold together?

#### DF with more than one Order

```python
df = all_data[all_data["Order ID"].duplicated(keep=False)]

df["Grouped"] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[["Order ID", "Grouped"]].drop_duplicates()

df
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)
```

### Question 5: What product sold the most? Why do you think it sold the most?

```python
grouped_product = all_data.groupby(all_data["Product"]).sum()['Quantity Ordered']
grouped_product.sort_values(ascending=False)
```

```python
grouped_prices = all_data.groupby(all_data["Product"]).mean()['Price Each']
grouped_prices
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(8, 6), sharex=False)

ax2 = ax1.twinx()

ax1.bar(grouped_product.index, grouped_product.values)
ax2.plot(grouped_prices.index, grouped_prices.values, color='g', alpha=0.5)

ax1.set_ylabel('Quantity Ordered', color='darkblue', fontsize=12)
ax1.set_xticklabels(grouped_product.index,rotation=90)
ax2.set_ylabel('Price [$]', color='g', fontsize=12)

plt.subplots_adjust(wspace=1, hspace=0)
```
